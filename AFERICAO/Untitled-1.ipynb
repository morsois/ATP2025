```python
# a) Contar ocorrências: defina uma função que recebe uma lista e um elemento como argumento e devolve
# o número de elementos na lista que são maiores que o elemento passado como argumento
def contarOcorrencias(lista, x):
    cont = 0
    for elem in lista:
        if x <= elem:
            cont += 1
    return cont

# b) Disjuntas?: defina uma função que recebe duas listas como argumento e devolve o valor booleano
# verdadeiro se as listas não tiverem nenhum elemento comum e falso caso contrário
def disjuntas(lis1, lis2):
    for elem1 in lis1:
        for elem2 in lis2:
            if elem1 == elem2:
                return False
    return True

# c) mergeListas: defina uma função que recebe duas listas de números ordenadas crescentemente como
# argumento e devolve uma nova lista ordenada de forma crescente com os elementos das duas listas.
def mergeListas(lis1, lis2):
    res = []; i = 0
    while i < len(lis1):
        j = 0
        while j < len(lis2):
            if lis1[i] <= lis2[j]:
                res.append(lis1[i])
                i += 1
            else:
                res.append(lis2[j])
                j += 1
    return res
```
